```python
#Balancing dataset using pycaret
```

```python
telecom=pd.read_csv("telecom_churn.csv")
```

```python
telecom.head()
```

```python
telecom.columns
```

```python
telecom=telecom[['International plan',
       'Voice mail plan', 'Number vmail messages', 'Total day minutes',
       'Total day calls', 'Total day charge', 'Total eve minutes',
       'Total eve calls', 'Total eve charge', 'Total night minutes',
       'Total night calls', 'Total night charge', 'Total intl minutes',
       'Total intl calls', 'Total intl charge', 'Customer service calls',
       'Churn']]
```

```python
telecom.head()
```

```python
import seaborn as sns
```

```python
sns.countplot(x="Churn",data=telecom)
```

Imbalanced dataset as ratio is 80:20 here so model will predict false for every case

```python
# Building PyCaret model without fixing the imbalance problem
exp=setup(data=telecom,target="Churn")
```

```python
compare_models()
```

```python
# Treating the dataset imablance problem using PyCaret
exp=setup(data=telecom,target="Churn",fix_imbalance=True)
```

```python
compare_models()
```

Problem solved now the predicted output is no longer imbalanced

```python
lgm=create_model("lightgbm")
```

```python
#tuning
tuned_lgm=tune_model(lgm)
```
